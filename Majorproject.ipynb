```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the pokemon file
pokemon = pd.read_csv("pokemon.csv")
```

```python
pokemon.shape
```

```python
pokemon.head()
```

```python
#checking if there are any null vakues in the dataset
pokemon.isnull().sum()
```

```python
y = pokemon['target'].values
X = pokemon.drop(columns=['target'], inplace=False).values
```

```python
y.shape
```

```python
X.shape
```

```python
X[0]
```

```python
#displaying the image of index 200 
plt.imshow(X[200].reshape(100,100,3))
```

```python
# importing from sklearn to split the model in train and test 
from sklearn.model_selection import train_test_split
```

```python
# keeping the size of test as 25% 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print( X_train.shape, y_train.shape )
```

```python
print( X_test.shape, y_test.shape)
```

```python
#importing the KNN algorithm
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=16)
```

```python
# model training 
model.fit(X_train, y_train)
```

```python
#predicting the y test values from x test values
y_pred = model.predict(X_test)
```

```python
#predicting the y prediction
y_pred
```

```python
# checking the accuracy for the test set
model.score(X_test, y_test)
```

```python
# printing for random x test indices
for i in [64, 55,70, 57, 75]:
    random_img = X_test[i]
    plt.imshow(random_img.reshape(100,100,3))
    plt.show()
    print("Actual Label : ",y_test[i])
    p = model.predict( [random_img] )
    print("Predicted Label : ", p)
```
